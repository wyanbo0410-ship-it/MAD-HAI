```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Data/literature_data_SDI.csv')

# generate training data set
comp_unique = np.unique(data[['C','Mn','Al','Si','Fe']],axis=0)

Z_unique = []
for i, comp in enumerate(comp_unique):
    tt = data[(data['C'] == comp[0]) & (data['Mn'] == comp[1]) & (data['Al'] == comp[2]) & (data['Si'] == comp[3])]
    z_mean = np.mean(tt['Z'])
    Z_unique.append(z_mean)
    
df = {'C':comp_unique[:,0], 'Mn':comp_unique[:,1], 'Al':comp_unique[:,2],'Si':comp_unique[:,3],'Fe':comp_unique[:,4]}
new_set = pd.DataFrame.from_dict(df)
new_set['Z'] = Z_unique
new_set.to_csv('Data/training_dataset.csv', index=False)

data_train = new_set
```
