```python
import pandas as pd
```

```python
import numpy as np
```

```python
def nan_to_element(a_list):
    a_c = a_list.copy()
    for i, a in enumerate(a_c):
        if np.isnan(a):
            a_c[i] = a_c[i-1]
    return a_c
```

```python
data_all = []
for i in range(8):
    data = pd.read_excel('Steel data-final-literature data-check.xlsx',sheet_name=i,header=2)
    data_all.append(data)
```

```python
data = data_all[0]
```

```python
C_list = []
Mn_list = []
Al_list = []
Si_list = []
Fe_list = []
UTS_list = []
for i in range(len(data_all)):
    data = data_all[i]
    C_list += list(nan_to_element(data['C']))
    Mn_list += list(nan_to_element(data['Mn']))
    Al_list += list(nan_to_element(data['Al']))
    Si_list += list(nan_to_element(data['Si']))
    Fe_list  +=list(nan_to_element(data['Fe']))
    UTS_list += list(nan_to_element(data['UTS (MPa)']))
```

```python
uni_elong_list = []
tot_elong_list = []
mask_uni_elong = []
mask_tot_elong = []
for i in range(len(data_all)):
    data = data_all[i]
    uni_elong = data['Uniform elongation (%)']
    uni_elong_list += list(uni_elong)
    tot_elong = data['Total elongation (%)']
    tot_elong_list += list(tot_elong)
    mask = np.ones(len(uni_elong),dtype=bool)
    for i,ele in enumerate(uni_elong):
        if isinstance(ele,str) or np.isnan(ele):
            mask[i] = False
    mask_uni_elong += list(mask)
    mask = np.ones(len(tot_elong),dtype=bool)
    for i,ele in enumerate(tot_elong):
        if isinstance(ele,str) or np.isnan(ele):
            mask[i] = False
    mask_tot_elong += list(mask)
```

```python
mask = mask_uni_elong

C_list = np.array(C_list)[mask]
Mn_list = np.array(Mn_list)[mask]
Al_list = np.array(Al_list)[mask]
Si_list = np.array(Si_list)[mask]
Fe_list = np.array(Fe_list)[mask]
UTS_list = np.array(UTS_list)[mask]
UE_list = np.array(uni_elong_list)[mask]
```

```python
df = {'C':C_list, 'Mn':Mn_list, 'Al':Al_list,'Si':Si_list, 'Fe':Fe_list,'UTS':UTS_list,'UE':UE_list}
```

```python
df = pd.DataFrame.from_dict(df)
```
